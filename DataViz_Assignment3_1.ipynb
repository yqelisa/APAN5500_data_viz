```python
# load dataset

df_full = pd.read_csv('economy.csv')
df_full.head()
```

```python
# count NAs

df_full.isnull().sum()  # Elisa: the dataset seems clean
```

```python
# dimension

print("The full dataset has {} rows and {} columns".format(df_full.shape[0],df_full.shape[1]))
df_full.shape
```

```python
# descriptive analysis

df_full.describe()

# Elisa: variables have different scale
```

```python
# histograms of variables

list_col_names = list(df_full.columns.values)

fig, axs = plt.subplots(4, 5, figsize=(15,8))

for idx, ax in enumerate(axs.ravel()):
    ax.hist(df_full[list_col_names[idx]])
    ax.set_title(list_col_names[idx])
    
plt.tight_layout()
```

```python
# time series

fig, axs = plt.subplots(4, 5, figsize=(15,8))

for idx, ax in enumerate(axs.ravel()):
    ax.plot(df_full[list_col_names[idx]])
    ax.set_title(list_col_names[idx])
    
plt.tight_layout()
```

```python
# scatter plot matrix

sns.set(style="ticks")
sns.pairplot(df_full.iloc[:,1:]) # exclude "Year"
```

```python
# correlation matrix

mat_correlations = df_full.iloc[:,1:].corr(method='pearson')
mat_correlations
```
